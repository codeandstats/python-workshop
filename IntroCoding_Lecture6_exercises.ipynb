#### Titanic data

```python
# Since pandas does not have any built in data, I am going to "cheat" and 
# make use of the `seaborn` library
import seaborn as sns 

titanic = sns. load_dataset('titanic')
titanic["3rdClass"] = titanic["pclass"]==3
titanic["male"] = titanic["sex"]=="male"

titanic
```

### Tasks:

- remove all rows which contain any missing values
- remove only those rows with missing age 
- replace the NAs in age with the mean age

```python
#you can remove all rows which contain any missing values:
titanic.dropna().shape
```

```python
titanic.dropna(subset="age")
```

```python
#in order to not lose all these rows you could replace the NAs:
meanAge = np.mean(titanic.age)
titanic.age.fillna(meanAge).tail()
```

```python
titanic.fillna({"age": meanAge}).tail()
```
